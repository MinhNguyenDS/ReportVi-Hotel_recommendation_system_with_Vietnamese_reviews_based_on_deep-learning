```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/DS310/datasetNLP/dataset2.5.csv')
```

```python
#Lựa chọn theo số lượng comment của Khách hàng
count_name = count_name = df['User_name'].value_counts().to_frame()
rank = np.array(count_name.index)
```

```python
df_rank = pd.DataFrame(rank)
df_rank[1] = df_rank.index + 1

rank_np = df_rank.to_numpy()
```

```python
df['User_name'] = df['User_name'].replace(rank_np[:,0], rank_np[:,1])
df = df.sort_values(by=['User_name'])
df['User_name'] = df['User_name'].replace(rank_np[:,1], rank_np[:,0])
```

```python
#dataset cần gán nhãn nếu có thể
df1 = df[0:4050]
```

```python
df2 = df[4050:df.shape[0]]
```

```python
df1 = df1.sample(frac= 1)
df2 = df2.sample(frac= 1)
```

```python
df1['User_name'].value_counts()
```

```python
df_new = pd.concat([df1, df2])
```

```python
df_new.to_csv('dataset2.5.csv', index=False, header=True)
```
