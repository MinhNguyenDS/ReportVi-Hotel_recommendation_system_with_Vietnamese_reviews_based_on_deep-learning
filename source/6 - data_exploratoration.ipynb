# Vẽ Word Cloud comment cho các nhãn

```python
Service = ''
Infrastructure = ''
Sanitary = ''
Location = ''
Attitude_0 = ''
Attitude_1 = ''
Attitude_2 = ''
for i in range(df.shape[0]):
  if df.Service[i] == 1: 
    Service += df.Comment[i]
  if df.Infrastructure[i] == 1: 
    Infrastructure += df.Comment[i]
  if df.Sanitary[i] == 1: 
    Sanitary += df.Comment[i]
  if df.Location[i] == 1: 
    Location += df.Comment[i]
  if df.Attitude[i] == 2: 
    Attitude_2 += df.Comment[i]
  if df.Attitude[i] == 1: 
    Attitude_1 += df.Comment[i]
  if df.Attitude[i] == 0: 
    Attitude_0 += df.Comment[i]
```

## Service

```python

wc = WordCloud(background_color="white", height=400, width=600)
wc.generate(Service)
wc.to_file("wordcloud_output.png")
img = cv2.imread('wordcloud_output.png', 0)
cv2_imshow(img)
```

# Điểm đánh giá

```python
df_rating = df.Rating.value_counts().sort_index(ascending = False).to_frame()
df_rating
```

```python
sum_10 = pd.DataFrame(df_rating.iloc[0:1].sum(), columns = ['Number'])
sum_10.rename(index={'Rating': '10.0'}, inplace=True)
sum_9 = pd.DataFrame(df_rating.iloc[1:9].sum(), columns = ['Number'])
sum_9.rename(index={'Rating': '9.0 - 9.9'}, inplace=True)
sum_8 = pd.DataFrame(df_rating.iloc[10:19].sum(), columns = ['Number'])
sum_8.rename(index={'Rating': '8.0 - 8.9'}, inplace=True)
sum_7 = pd.DataFrame(df_rating.iloc[20:25].sum(), columns = ['Number'])
sum_7.rename(index={'Rating': '7.0 - 7.9'}, inplace=True)
sum_6 = pd.DataFrame(df_rating.iloc[26:35].sum(), columns = ['Number'])
sum_6.rename(index={'Rating': '6.0 - 6.9'}, inplace=True)
sum_5 = pd.DataFrame(df_rating.iloc[36:45].sum(), columns = ['Number'])
sum_5.rename(index={'Rating': '5.0 - 5.9'}, inplace=True)
sum_4 = pd.DataFrame(df_rating.iloc[46:49].sum(), columns = ['Number'])
sum_4.rename(index={'Rating': '4.0 - 4.9'}, inplace=True)
```

```python
frame = [sum_10,sum_9,sum_8,sum_7,sum_6,sum_5,sum_4]
sum_all = pd.concat(frame)
sum_all
```

```python
sum_all.plot(kind = 'barh', ylabel = 'Number')
plt.show()
```
